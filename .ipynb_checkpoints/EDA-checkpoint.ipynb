We can see that almost half of the data has missing output variable, this tells us that the construciton or project is not completed yet

```python
model_df = df[df["Time Overrun (Months)"].notna()].copy()
```

```python
corr = model_df.corr(numeric_only=True)["Time Overrun (Months)"].sort_values(ascending=False)
corr
```

```python
important_num = corr[abs(corr) > 0.2]
important_num
```

```python
model_df.groupby("Mode")["Time Overrun (Months)"].mean().sort_values()
```

```python
cat_cols = model_df.select_dtypes("object").columns
```

```python
impact = {}

for col in cat_cols:
    if model_df[col].nunique() < 50:  # avoid high-cardinality columns
        impact[col] = model_df.groupby(col)["Time Overrun (Months)"].mean().std()

sorted(impact.items(), key=lambda x: x[1], reverse=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

X = model_df.select_dtypes(exclude="object").drop(columns=["Time Overrun (Months)"])
y = model_df["Time Overrun (Months)"]

model = RandomForestRegressor()
model.fit(X,y)

importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values(ascending=False)
```
